```python
def make_new_pdf(file_path):  
    """Input: a list of path files for PDFs  
    Output: a single PDF file"""  
    with open(file_path, mode='rb'):  
        #create and instance of both a pdf_reader & pdf_writer 
        pdf_reader = PyPDF2.PdfFileReader(file) 
        pdf_writer = PyPDF2.PdfFileWriter()  
        #count the number of pages in the file
        number_of_pages = pdf_reader.getNumPages()
        for page in range(number_of_pages):  
            #extract page information
            pdf_reader.getPage(page) 
            #write information to a new file
            pdf_writer.addPage(pdf_reader.getPage(page)) 
            #add the information to a new pdf file
            output = open('NewPages.pdf', 'wb')
            pdf_writer.write(output)
            output.close()
```

```python
def tokenize(song):
    joined_song = ' '.join(song)
    tokenized_song = word_tokenize(joined_song)
    
    return tokenized_song
```

```python
def count_vectorize(song, vocab=None):
    if vocab:
        unique_words = vocab
    else:
        unique_words = list(set(song))
    
    song_dict = {i:0 for i in unique_words}
    
    for word in song:
        song_dict[word] += 1
    
    return song_dict
```

```python
def term_frequency(BoW_dict):
    total_word_count = sum(BoW_dict.values())
    
    for ind, val in BoW_dict.items():
        BoW_dict[ind] = val/ total_word_count
    
    return BoW_dict
```
